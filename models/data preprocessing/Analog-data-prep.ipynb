```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('katanaloseis_clean.csv')
```

```python
df['EWS'] = pd.to_datetime(df['EWS']).dt.date
df['APO'] = pd.to_datetime(df['APO']).dt.date
df
```

```python
df['Average'] = df['KYB']/df['DAYS']
df
```

```python
months_df = pd.DataFrame()
months_df["School"] = []
months_df["Month"] = []
months_df["Monthly Consumption"] = []
months_df
```

```python
from datetime import datetime as dt

def get_months(n):

  lst = [dt.strptime('%2.2d-%2.2d' % (y, m), '%Y-%m').strftime('%b-%y') \
       for y in range(df['APO'][n].year, df['EWS'][n].year+1) \
       for m in range(df['APO'][n].month if y==df['APO'][n].year else 1, df['EWS'][n].month+1 if y == df['EWS'][n].year else 13)]
  
  return lst

month_timeseries = []
```

```python
import warnings
warnings.filterwarnings("ignore")
for ind in df.index:
  months = get_months(ind)
  i = 0
  start_days = df["APO"][ind].day
  end_days = df["EWS"][ind].day
  first_month = pd.Period(str(months[0])).days_in_month
  consumption = (first_month - start_days) * df["Average"][ind]
  df2 = {'School': df["ΔΙΕΥΘΥΝΣΗ"][ind], 'Month': months[0], 'Monthly Consumption': consumption}
  months_df = months_df.append(df2, ignore_index = True)
  for month in months[1:-1]:
    school = df["ΔΙΕΥΘΥΝΣΗ"][ind]
    month = month
    mnth = pd.Period(str(months[i])).days_in_month
    consumption = mnth * df["Average"][ind]
    df2 = {'School': school, 'Month': month, 'Monthly Consumption': consumption}
    months_df = months_df.append(df2, ignore_index = True)
    i = i+1
  consumption = end_days * df["Average"][ind]
  df2 = {'School': df["ΔΙΕΥΘΥΝΣΗ"][ind], 'Month': months[-1], 'Monthly Consumption': consumption}    
  months_df = months_df.append(df2, ignore_index = True)
```

```python
months_df['ID'] = months_df.groupby(['School'], sort=False).ngroup() + 1
```
