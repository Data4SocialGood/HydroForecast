```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import holidays
import os
%matplotlib inline
for filename in os.listdir("C:/Users/Nikolas/Downloads/NLOG_Data/NLOG_Data"):
    f = os.path.join("C:/Users/Nikolas/Downloads/NLOG_Data/NLOG_Data/", filename)
    print(f)
    if os.path.isfile(f):
        df = pd.read_excel(f, header=4)
        covid_months = ['2020-03', '2020-04', '2020-05', '2020-11', '2020-12', '2021-02', '2021-03', '2021-04']
        is_covid_mask = df['Record Date'].dt.strftime('%Y-%m').isin(covid_months)
        df['isCovid'] = is_covid_mask
        df['isChristmas'] = ((df['Record Date'].dt.month == 12) & ((df['Record Date'].dt.day >= 24) | (df['Record Date'].dt.day <= 31))) | ((df['Record Date'].dt.month == 1) & ((df['Record Date'].dt.day >= 1) & (df['Record Date'].dt.day <= 7)))
        df['isWeekday'] = df['Record Date'].dt.weekday < 5
        greek_holidays = holidays.GR(years=df['Record Date'].dt.year.unique())
        df['isHoliday'] = df['Record Date'].dt.date.astype('datetime64[ns]').isin(greek_holidays.keys())
        df['isSummer'] = ((df['Record Date'].dt.month == 6) & (df['Record Date'].dt.day >= 16)) | \
                 ((df['Record Date'].dt.month > 6) & (df['Record Date'].dt.month < 9)) | \
                 ((df['Record Date'].dt.month == 9) & (df['Record Date'].dt.day <= 10))
        df.to_csv(r"C:\Users\Nikolas\Desktop\NLOG_DATA_updated\{}.csv".format(filename))
```
